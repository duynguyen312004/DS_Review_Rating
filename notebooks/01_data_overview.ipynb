# 01 - Data Overview

Notebook nay dung de xem nhanh du lieu tho va tap trung vao 2 cot quan trong nhat:

- `Review`: noi dung binh luan, dung lam input cho model
- `Rating`: diem danh gia, dung de tao nhan sentiment

## Buoc 1 - Import thu vien

```python
import pandas as pd
from pathlib import Path
```

## Buoc 2 - Khai bao duong dan file du lieu

```python
PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_PATH = PROJECT_ROOT / "data" / "raw" / "restaurant_reviews.csv"

print("Project root:", PROJECT_ROOT)
print("Data path:", DATA_PATH)
print("File exists:", DATA_PATH.exists())
```

## Buoc 3 - Doc du lieu raw

```python
df = pd.read_csv(DATA_PATH)

print("So dong:", df.shape[0])
print("So cot:", df.shape[1])
```

## Buoc 4 - Xem nhanh du lieu goc

```python
df.head()
```

```python
df.columns
```

## Buoc 5 - Chi lay 2 cot can dung

```python
df_target = df[["Review", "Rating"]].copy()

df_target.head()
```

## Buoc 6 - Kiem tra thong tin 2 cot chinh

```python
df_target.info()
```

## Buoc 7 - Kiem tra missing values

```python
missing_summary = pd.DataFrame({
    "column": df_target.columns,
    "missing_count": df_target.isna().sum().values,
    "missing_percent": (df_target.isna().sum().values / len(df_target) * 100).round(2),
})

missing_summary
```

## Buoc 8 - Kiem tra cac gia tri Rating

```python
df_target["Rating"].value_counts(dropna=False)
```

## Buoc 9 - Tim Rating khong phai so

```python
rating_numeric = pd.to_numeric(df_target["Rating"], errors="coerce")

invalid_rating_rows = df_target[
    rating_numeric.isna() & df_target["Rating"].notna()
]

invalid_rating_rows
```

## Buoc 10 - Kiem tra review bi trung lap

```python
duplicate_review_count = df_target["Review"].duplicated().sum()

print("So review trung lap:", duplicate_review_count)
```
